```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2, 100)
```

```python
y = 2*x
```

```python
y2 = x**2
```

```python
plt.plot(x,y2, 'og')
plt.show()

```

```python
plt.plot(x,y, '*r')
plt.plot(x,y2, 'og')

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Titulo do gr√°fico')

plt.legend(['Vermelho', 'Verde'])
plt.show() # deixar o plt.show sempre no final da aba de comandos
```

```python
# criamos um grafico, na qual a linha horizontal recebe ate o valor maximo de x (2.0), e a linha vertical
# y, recebe at√© 4.0 (2x o valor de x), como foi dada a ordem para ser executada no comando

# plt.plot(y = 2*), logo apos executar o comando anterior, executamos este = (x,y, '*')

# e podemos observar que existem diversos valores em cima do grafico, e podemos retirar isso usando um cod
# que √© "plt.show()", embaixo do comando "plt.plot"

# podemos adicionar legendas ao grafico, usando o comando 'plt.xlabel'('escreva') (x = eixo, label = a linha)
# embaixo do ultimo comando, e replicar o comando para adicionar a legenda ao eixo y, trocando o comando
# xlabel por ylabel

# e para colocarmos um titulo no gr√°fico, utilizamos o comando plt.title('escreva o titulo')

# Podemos mudar o formato da tabela, mudando o que quisermos na linha de comando plt(plt.plot(x,y2, '.'))
# e o formato da linha da tabela sera feita por pontos

# podemos manipular a cor da nossa linha de grafico, usando ap√≥s a forma do grafico, a letra inicial da cor
# correspondente em que deseja usar (em ingles), por exemplo y:yellow, r:red
```
