```python
# Laço for em Estrutura de dados (listas, tuplas, dicionários, etc)

lista = [3, 8, 44, 100]
```

```python
for x in lista:
  print(x)
```

```python
# Podemos ver que ele pegou minha lista e printou ela

print(lista)
```

```python
# Mas sai de forma diferente da print normal, que ficaria assim

lista2 = [3, 8, 44, 100, 6, 9, 34]
```

```python
for i in lista2:
  if i%2 == 0:
    print(i)
```

```python
# Usando esse comando mesclado ao operador de módulo, podemos colocar para printar apenas numeros pares
# ou impares

for i in lista2:
  if i%2 == 1:
    print(i)
```

```python
# E mudando o numero sobra pra 1, printamos apenas os numeros impares
# também podemos mudar a operação para outro jeito que também funciona

for i in lista2:
  if i%2 != 0:
    print(i)
```

```python
# Colocando o comando de relação de "diferente", que quisemos dizer ao comando que é para imprimir
# os numeros que ao se dividirem por 2, o resultado fosse diferente de 0

lista3 = [[1,2,3],[2,3,5],[7,8,5,4,3,2]]
```

```python
for ii in lista3:
  print(ii)
```

```python
# Imprimimos diversas listas, como se fosse utilizado varios for diversas vezes, já que ele serve
# para que evitemos usar o print diversas vezes

# Também podemos imprimir os dicionários, mas devemos tomar um pouco mais de atenção

di = {'d1': 10, 'd2': 20, 'd3': 30, 'd4': 40, 'd5': 50,}
```

```python
di
```

```python
for i in di:
  print(i)
```

```python
# Ao executarmos o comando, na hora de printar, somente da o resultado das chaves, e nao dos valores

di.items()
```

```python
# Esses são nossos itens, e eles estão entre parenteses, entao estao armazenados como tuplas
# E para printar devemos fazer isso: for c(chaves), v(valor) in di: e em seguida print(c,v) e dessa forma
# printaremos as chaves e os valores relacionados as essas chaves

for c, v in di:
  print(c,v)
```

```python
# Pass, break e continue

for letra in 'Programador':
  if letra == 'a':
    pass
  print(letra)
```

```python
# Para percebermos a sutil diferença do pass e do continue

for letra in 'Programador':
  if letra == 'a':
    continue
  print(letra)
```

```python
# Deu para reparar na diferença entre esses dois comandos
# Agora o comando break, que o nome já quer dizer o que faz, "quebra", ele quebra a operaçao a partir do
# algoritmo predefinido

for letra in 'Programador':
  if letra == 'a':
    break
  print(letra)
```
