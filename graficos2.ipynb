```python
import numpy as np
import matplotlib.pyplot as plt

# funcoes

x = np.linspace(-1,1,100)
```

```python
y1 = 2*x
y2 = x**2
y3 = x**3
y4 = np.sin(x)
```

```python
fig, janela = plt.subplots(2,2) # para criar colunas com diferentes janelas, use esse comando atras dessa
# escrita (plt.subplots(nmr, nmr) ).

```

```python
# agora vamos começar a plotar os graficos, e para colocar dados na primeira janela, especificamos ela entre
# [] depois usamos o comando .plot e entre () colocamos as linhas, a cor e a forma da linha, como no exemplo

fig, janela = plt.subplots(2,2)
janela[0,0].plot(x,y1,'db')
janela[0,1].plot(x,y2, '*r')
janela[1,0].plot(x,y3, '_k')
janela[1,1].plot(x, y4, 'Hc')
fig.legend(['Reta', 'Rarabula', 'Cubica', 'Sin'])
plt.show()
```
